```python
#open groundtruth and compare line by line
werlist = []

file1 = 'C://Users//ngzhe//Desktop//CY2001-Research//Tesseract_2019_final.txt'
file2 = 'C://Users//ngzhe//Desktop//CY2001-Research//2019_COLLATED.txt'

with open(file1, encoding="utf8") as f1, open(file2, encoding="utf8") as f2:
    for line1, line2 in zip(f1, f2):
        wer = fastwer.score_sent(line1, line2, char_level=True)
        print(wer)
        werlist.append(wer)
```

```python

# Python program to get average of a list
def Average(lst):
    return sum(lst) / len(lst)

final_acc = Average(werlist)
```

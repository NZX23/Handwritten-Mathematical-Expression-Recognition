```python
#utility function from https://stackoverflow.com/a/10365472

def get_num(x):
    return int(''.join(ele for ele in x if ele.isdigit()))
```

```python
import os
import cv2

# Folder Path
path = "C://Users//ngzhe//Desktop//CY2001-Research//seshat//BACKUP Seshat//seshat-master//2019-SESHAT.txt"
#C:\Users\ngzhe\Desktop\CY2001-Research\seshat\BACKUP Seshat\seshat-master

times = []

for line in open(path, 'r', encoding='utf-8'): 
    if "milliseconds" in line:
        print(line)
        num = get_num(line)
        print(num)
        times.append(num)
```

```python
print(len(times))
```

```python
print(max(times))
print(min(times))
```

```python
#remove outliers
trimmed_list = times
trimmed_list.remove(max(trimmed_list))

print(len(trimmed_list))
print(max(trimmed_list))
```

```python
print(trimmed_list)
```

```python
#Plot histogram

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.hist(trimmed_list, density=False, bins=200, color = "darkslateblue")  # density=False would make counts
plt.ylabel('Amount of Data')
plt.xlabel('SESHAT Parser Computational Times (ms)');

plt.semilogy()
```

```python
#would boxplot be better?

fig = plt.figure(figsize =(10, 7))
 
# Creating plot
plt.boxplot(trimmed_list)
 
# show plot
plt.show()
```

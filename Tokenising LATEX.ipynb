```python
def longest_word(filename):
    with open(filename, 'r') as infile:
              words = infile.read().split()
    max_len = len(max(words, key=len))
    print(max_len)
    return [word for word in words if len(word) == max_len]

print(longest_word('COLLATED.txt'))
```

# Creating Tokens

```python
i=0
tokeniser = {}

#**NEED REMOVE \n in every key in file! dont append the \n as well!
with open("latex_vocab.txt", "r") as f:
    for line in f: 
        line = line.strip('\n')
        i+=1
        key, value = i, line
        tokeniser[key] = (value)
```

# Mapping Groundtruth to Tokens

```python
#so now tokeniser has the vocab encode in a dict distinctly

#and final_list has a list of lists of individual groundtruths tokenised

#so now need to map the final_list to tokeniser
```

```python
#tokeniser has key:value of digit:symbol, so to search by symbol need invert the dict

tokeniser = {v: k for k, v in tokeniser.items()}
```

```python
print(tokeniser)
```

```python
#for each list in final_list, get corresponding digit frm tokeniser, and append to encoded list?

tokenised_groundtruth = []

for list in final_list:
    encoded_list = []
    for item in list:
        corresponding_digit = tokeniser[item]
        if (isinstance(corresponding_digit, int)):
    
            encoded_list.append(corresponding_digit)
            
        else:
            print(item + "does not have corresponding digit. Corresponding digit is " + corresponding_digit)
    tokenised_groundtruth.append(encoded_list)
```

## Post Padding

```python
#find max length of list in list of lists

def find_max_list(list):
    list_len = [len(i) for i in list]
    print(max(list_len))

#print output#
find_max_list(tokenised_groundtruth)
```

```python
#10848, 53 will be final array size, pad the rest (at the end) with zero
```

```python
#initialise empty array, so that post padding will be settled
tokenised_groundtruth_arr = np.empty(shape=(10848, 53), dtype=int)

#for every list in tokenisied groundtruth list, overrite the corresponding dimension in the final arr
i=0
for list in tokenised_groundtruth:
    j=0
    for token in list:
        tokenised_groundtruth_arr[i][j] = token
        j+=1
        
    i+=1    
```

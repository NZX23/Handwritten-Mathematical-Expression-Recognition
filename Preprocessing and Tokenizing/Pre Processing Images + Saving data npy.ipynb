# Apply Preprocessing to every Image in Folder

```python
# run blur and erode and dilate on ALL pictures in dataset
# and save all the preprocessed images into a new file
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#from local_utils import detect_lp
%matplotlib inline
from matplotlib import gridspec
# from IPython.display import Image
import PIL.Image
import glob
from scipy import ndimage, misc
import os
import imageio
```

```python
outPath = "C://Users//ngzhe//Desktop//CY2001-Research//2019_IMGS_again"
path = "C://Users//ngzhe//Desktop//CY2001-Research//offline-crohme-master//2019_data_processed"
#"C://Users//ngzhe//Desktop//CY2001-Research//math-formula-recognition//data//test//2014 Images"

i=0
# iterate through the names of contents of the folder
for image_path in os.listdir(path):
    
    if ".png" in image_path: 
        # create the full input path and read the file
        input_path = os.path.join(path, image_path)
        img = cv2.imread(input_path)
        print(input_path)
        
        img = cv2.resize(img, (515, 200)) #, interpolation = cv2.INTER_AREA
        
        print(img.shape)

        img = cv2.medianBlur(img,3)
        img = cv2.erode(img,None,iterations = 1)
        

        # create full output path, 'example.jpg' 
        # becomes 'rotate_example.jpg', save the file to disk
        fullpath = os.path.join(outPath, image_path)

        imageio.imwrite(fullpath, img)
        i+=1

        print('{} processed'.format(fullpath))
    
print(i)
```

```python
#a = np.empty(shape=(10848, 310, 310, 3), dtype=int)

#a = np.empty(shape=(986, 256, 256, 3), dtype=int)

a = np.empty(shape=(11970, 389, 515, 3), dtype=int)

path = "C://Users//ngzhe//Desktop//CY2001-Research//2014_IMGS"
i=0

for image in os.listdir(path):
     if ".png" in image: 
            #convert image to numpy array
            image = os.path.join(path, image)
            img= PIL.Image.open(image)
            np_img = np.array(img)
            print(np_img.shape)

            #and add it to the ith index of a
            a[i] = np_img
            
            #where i will be incremented frm 0 to 10847
            i+=1
```
